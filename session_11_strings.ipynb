**Mutable vs Immutable**

- Mutable: charcters in a string can be changed by index operations

- Immutable: characters in a string can not be change index operations

```python
# strins are imuutable 
str1="welcome" # 'l' to 'L'
str1[2]='L'
```

```python
# list are mutable(cae be change)
l1=['w','e','l','c','o','m','e']
l1[2]='L'
l1
```
