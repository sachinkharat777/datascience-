**args-kwargs**

- args means we can provide multiple value without argument name

    - which means while calling function we can provide directly "virat",102
    - while define the function you no need to mention agrgument names also 
    - the output will cames in **tuple** format: ("virat",102)
- kwargs keyword arguments means we can provide multiple values with argument name
    - while define the function no need to provide the name of the arguments
    - while call the functions definitely we need to provide with argument names
    - name="virat", runs=102
    - the output comes in **dict** format: {'name':"virat",'runs':102}

```python
# with out name
# output in tuple format

def fun1(*args):
    print(args)
fun1(10,20)
```

```python
#with name
# output in dict format
def fun2(**kwargs):
    print(kwargs)
fun2(a=10,b=20)
```

```python
def fun2(**kwargs):
    print(kwargs)
fun2(name="sachin",age=20)
```
