**case-3: range(start,stop, step)**

- start value = start

- if step is postive sign
    - it is positive direction 

    - increment by **step valu**

    - last value = stop-1

- if step is negative sign
    - it is a negative direction(decrement)
    - decrement by **step value**
    - last value = stop+1

```python
for i in range(1,20,2):
    print(i, end=" ")

# step=1
# 
```

```python
import random
even_count=0
odd_count=0
even_sum=0
odd_sum=0
for i in range(5):
    num=random.randint(1,100)
    if num%2==0:
        print(f"even: {num}")
        even_count+=1
        even_sum+=num
    else:
        odd_count=1
        odd_sum+=num
        print(f"odd :{num}")
print(f" total count of even : {even_count}")
print(f" total count of odd : {odd_count}")
print(f" total sum of even : {even_sum}")
print(f" total sum of odd : {odd_sum}")
```
