```python
# wap ask the user take two number print subtraction multiplication and division
# ans: the subtraction of 100 and 200 is -100
#      the multiplication of 100 and 200 is 20000
#      the division of 100 and 200 is 0.5

n1=eval(input("enter first N1:"))
n2=eval(input("enter first N2:"))

add=round(n1+n2)
sub=round(n1-n2)
multi=round (n1*n2)
divi=round (n1/n2)

print(f"the addition of {n1} and {n2} is  : {add}")
print(f"the subscription of {n1} and {n2} is  : {sub}")
print(f"the multipliton of {n1} and {n2} is  : {multi}")
print(f"the division of {n1} and {n2} is  : {divi}")
```

```python
# wap ask the user take the radius find the area of circle
    # pi=3.14
    # r=100
   # area_of_circle=pi*r*r
radius=eval(input("enter radius of circle: "))
pi=3.14
area_of_circle= pi*radius*radius
print(f"the area of circle is : {area_of_circle}")
```

```python
# wap ask the user take height and breadth find the area of traingle
#       area_of_traingle=0.5*b*h
height=eval(input("enter a height:"))
breadth= eval(input("enter a breadth: "))
area_of_traingle=0.5*breadth*height
print(f"the area of traingle : {area_of_traingle}")
```

```python
# wap ask the user take 3 numbers find the average
n1=eval(input("enter n1:"))
n2=eval(input("enter n2:"))
n3=eval(input("enter n3:"))
avg= round((n1+n2+n3)/3,2)
print(f"the average of three numbers is : {avg}")
```
