**if-elif-else**

- when thre is one condition is there 
    - we can expect two answers:
        - in that first answer under **if** block 
        - second answer under **else** block
- when there are two conditons 
    - we can expect three answers:
        - in that first answer under **if** block based on conditon-1 
        - second answer under **elif** block based on condition-2
        - third answer under **else** block 

```python
one conditons ====> two answers ====> if-else
two conditions ====> 3 answers ====> if-elif-else
3 conditon      ======> 4 answers  =====> if-elif-elif-else

if elif need condition
```

```python
if <condition>:
    <write your code>
elif <conditon>:
    <write your code>
else:
    <write your code>
```

```python
# wap ask the user enter a number 
# if the number equal to 1 then print one
# otherwise if the number equal to 2 then print two
# otherwise if the number equal to 3 then print three
# otherwise enter a valid number 

n1=eval(input("enter a number"))
if n1== 1:
    print("one")
elif n1==2:
    print("two")
elif n1==3:
    print("three")
else:
    print("enter a valid number")
```
